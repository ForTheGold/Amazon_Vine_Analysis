```python
import pandas as pd
```

```python
df = pd.read_csv("vine_table_pg.csv")
df.head()
```

```python
rev_20_df = df.loc[df["total_votes"] >= 20]
rev_20_df.tail()
```

```python
helpful_votes = rev_20_df.loc[(rev_20_df["helpful_votes"]/rev_20_df["total_votes"]) >= 0.5]
helpful_votes.tail()
```

```python
vine_df = helpful_votes.loc[helpful_votes["vine"] == "Y"]
vine_df.head()
```

```python
no_vine_df = helpful_votes.loc[helpful_votes["vine"] == "N"]
no_vine_df.head()
```

```python
five_star_vote_count = len(helpful_votes.loc[helpful_votes["star_rating"] >= 5].index)
five_star_vine_vote_count = len(vine_df.loc[vine_df["star_rating"] >= 5].index)
five_star_not_vine_vote_count = len(no_vine_df.loc[no_vine_df["star_rating"] >= 5].index)

print("Total Number of Reviews: ", len(helpful_votes.index))
print("Total Number of 5 Star Reviews: ", five_star_vote_count)
print("Total Number of Vine Reviews: ", len(vine_df.index))
print("Total Number of Vine 5 Star Reviews: ", five_star_vine_vote_count)
print("Total Number of Nonvine Reviews: ", len(no_vine_df.index))
print("Total Number of Nonvine 5 Star Reviews: ", five_star_not_vine_vote_count)
print("Percent Vine 5 Star Reviews: ", round(((five_star_vine_vote_count / len(vine_df.index)) * 100), 2), "%")
print("Percent Nonvine 5 Star Reviews: ", round(((five_star_not_vine_vote_count / len(no_vine_df.index))* 100),2), "%")
```
